### Advanced Queries

### 1. Find the average vote count for movies with a vote average greater than 8.

```python
query = """ select avg(vote_count) from Top_rated_movies where vote_average>8.5"""
cursor.execute(query)
data = cursor.fetchall()
data[0][0]
```

### 2. Retrieve the most popular movie in each year.


```python
query = """ select m1.title , year(m1.release_date) as year , m1.popularity from Top_rated_movies m1 join (select  year(release_date) as year , max(popularity) as max_p from Top_rated_movies group by year) m2 on year(m1.release_date) = m2.year and m1.popularity = m2.max_p """
cursor.execute(query)
data = pd.DataFrame(cursor.fetchall())
data
```
